```python
# -----------------------------
# IMPORTS
# -----------------------------
import string

# -----------------------------
# LANGUAGE DETECTION KEYWORDS
# -----------------------------
language_keywords = {
    "hindi": ["है","यह","बहुत","और"],
    "telugu": ["ఇది","చాలా","మరియు"],
    "tamil": ["இது","மிகவும்","மற்றும்"],
    "kannada": ["ಇದು","ತುಂಬಾ","ಮತ್ತು"],
    "malayalam": ["ഇത്","വളരെ","കൂടാതെ"]
}

# -----------------------------
# BIG SENTIMENT DICTIONARY
# -----------------------------
sentiment_dict = {

    "hindi": {
        "positive": [
            "अच्छा","बहुत अच्छा","शानदार","बेहतरीन","उत्तम","खुश","पसंद",
            "लाजवाब","सुंदर","संतुष्ट","कमाल","मज़ेदार","प्रसन्न",
            "आनंद","श्रेष्ठ","उत्कृष्ट","बढ़िया","जबरदस्त","दिलचस्प"
        ],
        "negative": [
            "बुरा","बहुत बुरा","खराब","बेकार","नापसंद","घटिया",
            "निराश","दुखी","खराबी","घिनौना","भयानक",
            "असहनीय","निराशाजनक","कमजोर","गंदा","बेहद खराब"
        ]
    },

    "telugu": {
        "positive": [
            "మంచి","చాలా మంచి","అద్భుతం","బాగుంది","ఇష్టం",
            "సంతోషం","సంతృప్తి","అత్యుత్తమం","శ్రేష్ఠం",
            "అద్భుతమైన","ఆనందం","బాగున్నది"
        ],
        "negative": [
            "చెడు","చాలా చెడు","బాగా లేదు","నాసిరకం",
            "చెత్త","అసహ్యం","నిరాశ","దారుణం",
            "పొరపాటు","చెడ్డది","తక్కువ"
        ]
    },

    "tamil": {
        "positive": [
            "நல்ல","மிக நல்ல","அருமை","சிறந்தது",
            "அற்புதம்","மகிழ்ச்சி","திருப்தி","சூப்பர்",
            "சிறப்பு","மிக சிறந்த"
        ],
        "negative": [
            "மோசம்","மிக மோசம்","கெட்ட","பழுதான",
            "அருவருப்பு","வெறுப்பு","நிராசை","தவறு"
        ]
    },

    "kannada": {
        "positive": [
            "ಒಳ್ಳೆಯದು","ಬಹಳ ಒಳ್ಳೆಯದು","ಅದ್ಭುತ","ಚೆನ್ನಾಗಿದೆ",
            "ಇಷ್ಟ","ಸಂತೋಷ","ತೃಪ್ತಿ","ಅತ್ಯುತ್ತಮ","ಶ್ರೇಷ್ಠ"
        ],
        "negative": [
            "ಕೆಟ್ಟದು","ಬಹಳ ಕೆಟ್ಟದು","ಕೆಟ್ಟ","ಹಾಳಾಗಿದೆ",
            "ಅಸಹ್ಯ","ನಿರಾಶೆ","ದುಃಖ","ಭಯಾನಕ"
        ]
    },

    "malayalam": {
        "positive": [
            "നല്ലത്","വളരെ നല്ലത്","മികച്ചത്","അദ്ഭുതം",
            "ഇഷ്ടം","സന്തോഷം","തൃപ്തി","ശ്രേഷ്ഠം"
        ],
        "negative": [
            "മോശം","വളരെ മോശം","കേട്","വെറുപ്പ്",
            "നിരാശ","ദോഷം","അസഹ്യം"
        ]
    }
}

# -----------------------------
# LANGUAGE DETECTION
# -----------------------------
def detect_language(text):
    for lang, keywords in language_keywords.items():
        if any(word in text for word in keywords):
            return lang
    return "unknown"

# -----------------------------
# SENTIMENT ANALYSIS
# -----------------------------
def analyze_sentiment(text):
    
    text = text.lower()
    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )
    
    language = detect_language(text)
    
    if language == "unknown":
        print("❌ Language not detected")
        return
    
    pos_count = 0
    neg_count = 0
    
    # Phrase-Based Matching
    for phrase in sentiment_dict[language]["positive"]:
        if phrase in text:
            pos_count += 1
    
    for phrase in sentiment_dict[language]["negative"]:
        if phrase in text:
            neg_count += 1
    
    if pos_count > neg_count:
        sentiment = "Positive"
    elif neg_count > pos_count:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    print("\nDetected Language:", language)
    print("Positive Matches:", pos_count)
    print("Negative Matches:", neg_count)
    print("Sentiment:", sentiment)

# -----------------------------
# USER INPUT
# -----------------------------
text = input("Enter sentence: ")
analyze_sentiment(text)
```
